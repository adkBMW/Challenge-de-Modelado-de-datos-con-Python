### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
from google.colab import sheets
sheet = sheets.InteractiveSheet(df=tienda)
```

#1. Análisis de facturación



```python
ingresos = {
    "Tienda 1": tienda["Precio"].sum(),
    "Tienda 2": tienda2["Precio"].sum(),
    "Tienda 3": tienda3["Precio"].sum(),
    "Tienda 4": tienda4["Precio"].sum()
}

df_ingresos = pd.DataFrame(list(ingresos.items()), columns=["Tienda", "Ingreso Total"])
df_ingresos
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.bar(df_ingresos["Tienda"], df_ingresos["Ingreso Total"], color="skyblue")
plt.title("Ingresos Totales por Tienda")
plt.ylabel("Ingreso Total ($)")
plt.show()
```

# 4. Productos más y menos vendidos

```python
# Productos más y menos vendidos en Tienda 1
productos_t1 = tienda["Producto"].value_counts()
print("Tienda 1 - Top 5 más vendidos:")
print(productos_t1.head())
print("\nTienda 1 - 5 menos vendidos:")
print(productos_t1.tail())

# Productos más y menos vendidos en Tienda 2
productos_t2 = tienda2["Producto"].value_counts()
print("\nTienda 2 - Top 5 más vendidos:")
print(productos_t2.head())
print("\nTienda 2 - 5 menos vendidos:")
print(productos_t2.tail())

# Productos más y menos vendidos en Tienda 3
productos_t3 = tienda3["Producto"].value_counts()
print("\nTienda 3 - Top 5 más vendidos:")
print(productos_t3.head())
print("\nTienda 3 - 5 menos vendidos:")
print(productos_t3.tail())

# Productos más y menos vendidos en Tienda 4
productos_t4 = tienda4["Producto"].value_counts()
print("\nTienda 4 - Top 5 más vendidos:")
print(productos_t4.head())
print("\nTienda 4 - 5 menos vendidos:")
print(productos_t4.tail())
```
